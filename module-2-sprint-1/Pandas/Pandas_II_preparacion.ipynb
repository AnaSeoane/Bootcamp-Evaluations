### Ejercicios Preparación:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/Spotify_Youtube.csv', index_col='Unnamed: 0')

pd.options.display.max_columns = None
df.head(1)
```

- Los nombres de las columnas empiezan con mayúsculas, poned todos los nombres de las columnas en minúsculas.

```python
# hacemos una list comprenhension para crear un diccionario que nos permita modificar cada columna ({nombre columna actual: nueva columna}):
nuevas_columnas = {col: col.lower() for col in df.columns}

# renombramos las columnas del dataframe:
df.rename(columns = nuevas_columnas, inplace = True)

df.head(1)
```

___
- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
# comprobamos el tipo de todas ellas:
df.loc[:, 'danceability':'instrumentalness'].dtypes
```

    -> Como imaginábamos, eran `string` (con comas no podrían haber estado como `float`)

```python
def modificar_elementos(columna, elemento_modificar, elemento_nuevo):
    ''' Esta función corrige elementos dentro de los valores de una columna.
        Recibe tres parámetros:
            - columna: indica la columna a modificar
            - elemento_modificar: indica cuál es el elemento a reemplazar
            - elemento_nuevo: indica cuál es el elemento nuevo a indicar'''
    df[columna] = df[columna].replace(elemento_modificar, elemento_nuevo, regex=True)
```

```python
columnas_modificar = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

for column in columnas_modificar:
    modificar_elementos(column, ',', '.')
```

```python
df.head(2)
```

- Después de haber hecho los cambios, chequead los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, cread otra función o modificar la anterior para que sean de tipo float.

```python
df.loc[:, columnas_modificar].dtypes
```

    -> Como era de esperar, aunque habíamos modificado la coma por el punto, el tipo seguía siendo `string`.

```python
df[columnas_modificar] = df[columnas_modificar].astype(float)
```

```python
df.loc[:, columnas_modificar].dtypes
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "title" y "album".

```python
# Creamos una función que modifique la coma por el punto:
def minusc(nombre):
    """Esta función modifica los valores y los convierte a todo en minúsculas.
        Intentamos ejecutar el cambio y, si el valor es nulo, devuelve nulo"""
    try:
        return nombre.lower()
    except:
        return np.nan
```

```python
columnas_modificar2 = ['track', 'title', 'album']

for column in columnas_modificar2:
    df[column] = df[column].apply(minusc)
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
# Primero vamos a ver los ejemplos que lo contienen:
df[df['artist'].str.contains('\$')]['artist'].value_counts()
```

    -> Vistos los nombres que contienen el símbolo '$', realmente son un juego con las letras (reemplazando a las normales 'S'). La función que creemos será un cambio de '$' por 'S'.

```python
#Volvemos a llamar a la función que creamos al principio (donde reemplazamos ',' por '.':
modificar_elementos('artist', '\$', 'S')
```

```python
# Comprobamos que ahora sí esté correcto:
df[df['artist'].str.contains('\$')]['artist'].count()
```
