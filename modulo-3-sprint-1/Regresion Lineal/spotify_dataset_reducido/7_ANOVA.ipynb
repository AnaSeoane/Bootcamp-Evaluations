```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import RobustScaler
from scipy.stats import kstest

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_pickle('data/spotify_songs_estandarizado.pkl')
df.head(2)
```

```python
columnas_anova = "popularity ~ "

for i in range(len(df.columns)):
    if df.columns[i] != "popularity":

        if i == len(df.columns)-1:
            columnas_anova = columnas_anova + df.columns[i]
        
        else:
            columnas_anova = columnas_anova + df.columns[i] + " + "
            i += 1
    
    else:
        i+=1
        
columnas_anova
```

```python
lm = ols(columnas_anova, data=df).fit()
```

```python
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

En el supuesto en el que pudiesemos usar ANOVA en este caso, las conclusiones ser铆an:

1. Nuestra VR ("popularity") estar铆a explicada en un 2% por las variables predictoras.
2. En concreto, las variables que predice un poco mejor nuestra VR son: "explicit", "time_signature", "danceability", "energy", "loudness", "speachness", "instrumentalness", "valence" y "tempo" (p-value < 0.05)
